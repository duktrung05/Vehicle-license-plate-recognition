# 🏋️ Training Analysis

Analyze YOLOv8 training results.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('dark_background')
from pathlib import Path

# Load training results if available
results_path = Path('../models/yolov8/results.csv')

if results_path.exists():
    df = pd.read_csv(results_path)
    df.columns = df.columns.str.strip()
    print('Columns:', df.columns.tolist())
    print(df.tail())
else:
    print('No results.csv found. Train a model first.')
    print('Command: yolo train model=yolov8n.pt data=data/annotations/data.yaml epochs=100')
```

```python
# Plot training curves
if results_path.exists():
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    fig.suptitle('YOLOv8 Training Curves', fontsize=16, fontweight='bold')

    metrics = [
        ('train/box_loss', 'Train Box Loss', 'steelblue'),
        ('train/cls_loss', 'Train Class Loss', 'orange'),
        ('val/box_loss', 'Val Box Loss', 'green'),
        ('metrics/precision(B)', 'Precision', 'cyan'),
        ('metrics/recall(B)', 'Recall', 'magenta'),
        ('metrics/mAP50(B)', 'mAP@0.5', 'yellow'),
    ]

    for ax, (col, title, color) in zip(axes.flat, metrics):
        if col in df.columns:
            ax.plot(df.index, df[col], color=color, linewidth=2)
            ax.set_title(title, fontsize=11)
            ax.set_xlabel('Epoch')
            ax.grid(alpha=0.3)
        else:
            ax.text(0.5, 0.5, f'{col}\nnot found', ha='center', va='center', transform=ax.transAxes)
            ax.set_title(title)

    plt.tight_layout()
    plt.savefig('../reports/figures/training_curves.png', dpi=150, bbox_inches='tight')
    plt.show()
```
